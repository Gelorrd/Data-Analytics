```python
def check_price():
    # Connect to Website and pull in data

    URL = 'https://www.amazon.com/dp/B0DKMXR53N'
    
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9"
    }
    
    page = requests.get(URL, headers=headers)
    
    soup = BeautifulSoup(page.content, 'html.parser')
        
    # Find the product title
    title = soup.find(id='productTitle').get_text(strip=True)
    
    # Find the product price
    whole = soup.find('span', class_='a-price-whole').get_text().replace(',', '').strip().rstrip('.')
    fraction = soup.find('span', class_='a-price-fraction').get_text().strip()
    price = float(f"{whole}.{fraction}")
    
    
    print(f"{title}")
    print(price)  
    
    # Checking Price Change in Percentage
    file_name = r"C:\Users\angel\AmazonPriceCheckerDatasetDataset.csv"
    previous_price = None

    if os.path.exists(file_name):
        with open(file_name, 'r', encoding='UTF8') as f:
            reader = list(csv.reader(f))
            if len(reader) > 1:
                previous_price = float(reader[-1][1])
    
    if previous_price is not None:
        percent_change = round(((price - previous_price) / previous_price) * 100, 2)
    else:
        percent_change = 0.0
        
    print(percent_change)
    
    today = datetime.date.today()

    # Downloading csv file

    header = ['Title', 'Price', 'Date', 'Percent_Change']
    data = [title, price, today, percent_change]
    
    with open('AmazonPriceCheckerDataset.csv', 'a+', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)    
        
```

```python
# Check the price every 24hrs (86,400 secs)
while(True):
    check_price()
    time.sleep(86400)
```
